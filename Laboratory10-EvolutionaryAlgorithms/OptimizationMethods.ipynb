
## Optimization problems if D is continous:
- Sphere function (unimodal): f(x)=(x1)^2+(x2)^2+...+xn^2 = sum(x^2)
- Griewank function( uni-modal,non-convez function): f(x)=1+1/4000*sum(x^2)-prod(cos(x/sqrt(i)))
- Rastrigin function (multi-modal): f(x)=10n+sum(x^2-10cos(2*pi*x))



```python
def randomUniformVector(MIN=-5,MAX=5,D=1):
    return [[generateNewValue(MIN,MAX) for i in range(D)] for j in range(0,1000)]
def randomUniformVector2D(MIN=-5,MAX=5,D=2,S=10):
    return [np.linspace(MIN,MAX,S) for i in range(D)]
```

## Sphere function

```python
def sphere(x):
    return sum(xi**2 for xi in x)
plot1DOptimisation(randomUniformVector(),sphere)
```

## Griewank function

```python
def griewank(x):
    return 1 + sum(xi**2/4000 for xi in x) - np.prod([np.cos(xi/np.sqrt(i+1)) for i,xi in enumerate(x)])
plot1DOptimisation(randomUniformVector(),griewank)
```

## Rastrigin function

```python
def rastrigin(x):
    return 10*len(x) + sum(xi**2 - 10*np.cos(2*np.pi*xi) for xi in x)
plot1DOptimisation(randomUniformVector(),rastrigin)
```
